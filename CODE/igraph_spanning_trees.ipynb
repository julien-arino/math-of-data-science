```r
library(igraph)
```

Let us make some graphs and play with spanning trees..

First, a complete graph.

```r
G = make_full_graph(10)
plot(G)
```

```r
c = components(G)
c$no
```

Single connected component, the graph is connected. Let us find a spanning tree. Recall that spanning trees are far from unique, so here we use the minimum spanning tree (function `minimum.spanning.tree` or `mst` for short - see the help [here](https://igraph.org/r/html/latest/mst.html)): not unique either, but at least, more predictable.

```r
st = minimum.spanning.tree(G)
plot(st)
```

Yes, we have ourselves a tree! Let us now look at the directed case,

```r
G = make_full_graph(5, directed = TRUE)
plot(G)
c = components(G)
c$no
```

Of course, this is still connected. Recall that by default, `components` checks for "weak" components, i.e., components when the direction of arrows does not count, i.e., like the cycles we looked at in class.

```r
st = minimum.spanning.tree(G)
plot(st)
```

The minimum spanning tree has got rid of double arrows, since that would be cycles.
